```python
def is_palindrome(N):
    original = N
    reverse = 0
    while N != 0:
        last_digit = N % 10
        reverse = (reverse * 10) + last_digit
        N = N // 10
    return original == reverse


N = int(input())
if is_palindrome(N):
    print("true")
else:
    print("false")
```
