```python
def even_odd_digits(N):
    even_sum = 0
    odd_sum = 0
    while N != 0:
        last_digit = N % 10
        if last_digit % 2 == 0:
            even_sum += last_digit
        else:
            odd_sum += last_digit
        N = N // 10
    return even_sum, odd_sum


N = int(input())
even_sum, odd_sum = even_odd_digits(N)
print(even_sum, odd_sum)
```
